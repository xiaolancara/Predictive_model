```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (10,8)
```

```python
def PI(x,B):
    pi = 1/(1+np.exp(np.dot(x,B)))
    return pi
def log_likelihood(x,y,B):
    log_likelihood = np.sum(np.log((np.exp(np.dot(x,B))*PI(x,B))**y)+(np.log(PI(x,B)**(1-y))))
    return log_likelihood


# set seed to random state
np.random.seed(10)

# generate x, y, and B
n=100
x = np.random.uniform(size=(n,))
y = 1*(x+0.1*np.random.normal(size=(n,))>0.5)
Beta = np.arange(-4,4,0.0001)



# set initial lnL_B negative infinity for further comparation 
pre_lnL_B = -np.inf
lnL_B= []
for B in Beta:
    curr_lnL_B = log_likelihood(x,y,B)
    lnL_B.append(curr_lnL_B)
    if curr_lnL_B > pre_lnL_B:
            pre_lnL_B = curr_lnL_B
            Best_beta = B
            
# print the result with Maximum log likelihood and Best beta
print("Maximum log likelihood: ", pre_lnL_B)
print("Best beta: ", Best_beta)

# plot lnL vs. B
plt.style.use('ggplot')
plt.plot(Beta,lnL_B)
plt.xlabel('Beta')
plt.ylabel('Log likelihood')
plt.title('Beta - Maximize Log likelihood')
```

From the above result and graph, I'll choose B = 1.5748 as the estimate of beta, the parameter of the model that has the maximum Log likelihood in the logistic regression.
