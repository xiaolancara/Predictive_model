# 1  (1)

Complete the function `estimate(x, y, x0, h)` shown below:



The function should return the estimate $\hat{Y} (x_0)$.

```python
def estimate(x, y, x0, h):
    w = np.exp( -(x-x0)**2 / h )
    W = np.diag(w)
    x = x[:,None]
    X = np.concatenate(
        (np.ones(shape=x.shape),
        (x-x0)), axis=1
    )
    B = np.linalg.inv(X.T @ W @ X) @ (X.T @ W @ y)
    return B[0]
```
