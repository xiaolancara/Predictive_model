### 2 (a) Show analytically that the sample mean is the value of mu that minimizes SSE


### 2(b) Plot SSE vs mu_candidates

```python
import random
import numpy as np
from matplotlib import pyplot as plt

n_samples = 10
index = n_samples / 2
samples = []
mu = 1
# generate sample of y
while index:
    index -= 1
    random_number = random.randrange(-1,1)
    # append two samples in each loop that the mean of two samples is equal to 1
    samples.append(random_number)
    samples.append(mu*2-random_number)

print('sample', samples)

# verify the mu of samples is 1
mu = sum(samples) / len(samples)
print('mu', mu)

def SSE(mean):
    list_error = [(sample-mean)**2 for sample in samples]
    print('error', list_error)
    print('---------------------------')
    SSE = sum(list_error)
    return SSE
mu_candidates=np.arange(0, 2.01, .1)
list_SSE = [SSE(mu) for mu in mu_candidates]
print('list_SSE', list_SSE)

# plot the relationship between mu and SSE
plt.plot(mu_candidates, list_SSE, c='green')
plt.xlabel('mu')
plt.ylabel('SSE')
```

From the above plot, it concludes that when mu = 1, it has lowest SSE. Thus the value of mu that minimizes SSE.

### 3. D2.1 (a) Draw a scatterplot of Acres/Gardener (Y ) versus Consumers/Gardener (X)

```python
import pandas as pd
pd.options.mode.chained_assignment = None

df = pd.read_csv('https://socialsciences.mcmaster.ca/jfox/Books/Applied-Regression-3E/datasets/Sahlins.txt', delim_whitespace=True)
df
```

```python
import seaborn as sns
sns.regplot(x=df.consumers, y=df.acres)
```

From the graph, there's a positive relationship between consumers and acres, it's linear and weak. And yes, we can see there're several households have unusual distribution such as when consumers is 1.2, the acres is over than 3 and when consumer equals to 1.6, the acres has only 1.25 etc.

### 3. D2.1 (b) Divide the 20 households into three groups and calculate mean then plot

```python
# divide into 3 groups
first_group = df[:7]
second_group = df[7:13]
third_group = df[-7:]
display(first_group)
display(second_group)
display(third_group)
```

```python
mean_consumers = []
mean_acres = []
for group in [first_group,second_group,third_group]:
    mean_consumers.append(group.mean(axis = 0).consumers)
    mean_acres.append(group.mean(axis = 0).acres)
sns.regplot(x=mean_consumers, y=mean_acres)
plt.xlabel('consumers')
plt.ylabel('acres')
```

In this three group mean value gragh, the values are almost located on the regression line that shows there's a positive relationship between consumers and acres.

### 3. D2.1 (c)  Recalculate the means in groups 1 and 2 omitting the outlying observations, and redraw the nonparametric regression line

```python
# drop the outlying observation in first group
first_group.drop(index=4, inplace=True)
second_group.drop(index=12, inplace=True)
display(first_group)
display(second_group)
mean_consumers = []
mean_acres = []
for group in [first_group,second_group,third_group]:
    mean_consumers.append(group.mean(axis = 0).consumers)
    mean_acres.append(group.mean(axis = 0).acres)
sns.regplot(x=mean_consumers, y=mean_acres)
plt.xlabel('consumers')
plt.ylabel('acres')
```

After omitting the outlying observations, the relationship between consumers and acres is strongly positive, with all values closer to the regression line.
