```python
import numpy as np
from statsmodels.graphics.tsaplots import plot_acf
import matplotlib.pyplot as plt
%matplotlib inline
```

# Part 1

```python
# Simulate a sample of 100 elements from the ARMA(1,1) process
np.random.seed(1)
n=100
phi = [0.6,0.6,-0.8,0.5]
theta = [0.4,-0.4,0.55,-0.9]

for i in range(len(phi)):
    y = [1]
    residuals = [np.random.normal(0, 1)]
    for _ in range(1,n):
        residuals.append(np.random.normal(0,1))
        y.append(phi[i]*y[-1]+ residuals[-1]+theta[i]*residuals[-2])
    y = np.array(y).reshape(-1,1)
    plot_acf(y)
    plt.title(f'phi = {phi[i]}  theta = {theta[i]}')
    plt.show()
```

From above auto correlation results. It shows that :
- when φ = 0.6, θ = 0.4, lag1 is good for the order in time series prediction.
- when φ = 0.6, θ = −0.4, none of lag is good for the order in time series predicting.
- when φ = −0.8, θ = 0.55, lag1 is good for the order in time series predicting.
- when φ = 0.5, θ = −0.9, lag1 is good for the order in time series predicting.

These results prove that we used the MA(1).
